```
##Importing Necessary Library
import librosa
import torch
import tensorflow as tf
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

**Initialising Model**

```
#load pre-trained model and tokenizer
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

**Converting Audio to text**

```
##Audio file path
path="audio.mp3"
```

```
#Display the audio
import IPython.display as display
display.Audio(path, autoplay=True)
```

```
#load any audio file of your choice
speech, rate = librosa.load(path,sr=16000)

#tokenizing
input_values = tokenizer(speech, return_tensors = 'pt').input_values

#Store logits (non-normalized predictions)
logits = model(input_values).logits

#converting pytorch tensor to numpy
logits_np = logits.detach().numpy()

#Store predicted id's
predicted_ids = tf.argmax(logits_np, axis=-1)

#decode the audio to generate text
transcriptions = tokenizer.decode(predicted_ids[0])

input=transcriptions.lower()
```

```
#spell correction
import textblob
from textblob import TextBlob

blob=TextBlob(input)

input=str(blob.correct())
```

**Translating Audio to Hindi Language**

```
##Importing Fine-tuned Model
from transformers import pipeline

pipe = pipeline("text2text-generation", model="Abhaygpt04/text_trans_en_hi")
```
